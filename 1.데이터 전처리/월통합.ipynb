```python
import pandas as pd

# 첫 번째 파일을 기준으로 병합을 시작
df_month_merged = pd.read_csv(f"/content/drive/MyDrive/workspace/kaggle&github/~09.30_빅콘테스트/data/제주_관광_데이터/제주 관광수요 예측 데이터_비짓제주 월별 데이터/JT_MT_ACCTO_TRRSRT_SCCNT_LIST_202301.csv", encoding='utf-8')[['CL_CD','CL_NM','AREA_NM','ADDR','BASE_YEAR','JAN_VIEWS_CO']]

# 나머지 2월부터 12월까지의 파일을 불러와 필요한 열만 병합
month_columns = ['FEB_VIEWS_CO', 'MAR_VIEWS_CO', 'APR_VIEWS_CO', 'MAY_VIEWS_CO', 'JUN_VIEWS_CO',
                 'JULY_VIEWS_CO', 'AUG_VIEWS_CO', 'SEP_VIEWS_CO', 'OCT_VIEWS_CO', 'NOV_VIEWS_CO', 'DEC_VIEWS_CO']

for month, col in zip(range(2, 13), month_columns):
    df_temp = pd.read_csv(f"/content/drive/MyDrive/workspace/kaggle&github/~09.30_빅콘테스트/data/제주_관광_데이터/제주 관광수요 예측 데이터_비짓제주 월별 데이터/JT_MT_ACCTO_TRRSRT_SCCNT_LIST_2023{str(month).zfill(2)}.csv", encoding='utf-8')[['CL_CD','CL_NM','AREA_NM','ADDR','BASE_YEAR', col]]
    df_month_merged = pd.merge(df_month_merged, df_temp, on=['CL_CD','CL_NM','AREA_NM','ADDR','BASE_YEAR'], how='outer')

# 결과 확인
print(df_month_merged.head())

# merge된 데이터 저장
df_month_merged.to_csv('/content/drive/MyDrive/workspace/kaggle&github/~09.30_빅콘테스트/data/제주_관광_데이터/제주 관광수요 예측 데이터_비짓제주 월별 데이터/month_total.csv')
```
