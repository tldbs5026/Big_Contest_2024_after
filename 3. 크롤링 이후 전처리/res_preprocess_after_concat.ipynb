```python
import ast
import json


# 특징: 한글로 디코딩
def decode_features(feature_column):
    try:
        # 입력값이 문자열인지 확인
        if isinstance(feature_column, str):
            # 문자열 데이터를 JSON 형식으로 정제
            sanitized_column = feature_column.replace('""', '"').replace('", "', '", "')
            # 문자열을 Python 리스트로 변환
            feature_list = ast.literal_eval(sanitized_column)
            # 각 항목의 첫 번째 요소를 디코딩
            return ", ".join([json.loads(f'"{item[0]}"') for item in feature_list if item[0]])
        else:
            return ""  # 문자열이 아니면 빈 문자열 반환
    except (ValueError, SyntaxError, TypeError):
        return ""

# '특징' 필드 변환
data['메뉴'] = data['메뉴'].apply(decode_features)
data['특징'] = data['특징'].apply(decode_features)
```

```python
# 사진: 첫 번째 URL만 추출
def extract_first_image(photo_column):
    try:
        # 문자열을 튜플 형태로 변환
        photo_list = ast.literal_eval(photo_column)
        # 첫 번째 URL만 반환
        return photo_list[0]
    except (ValueError, SyntaxError, IndexError):
        return ""

# '사진' 필드 변환
data['사진'] = data['사진'].apply(extract_first_image)
```
