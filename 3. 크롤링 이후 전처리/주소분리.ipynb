```python
def extract_first_url(photo_column):
    try:
        # 문자열을 튜플 형태로 변환 후 첫 번째 URL만 반환
        urls = eval(photo_column) if isinstance(photo_column, str) else []
        return urls[0] if urls else ""
    except (SyntaxError, ValueError, IndexError):
        return ""

tour_crawling['사진'] = tour_crawling['사진'].apply(extract_first_url)
tour_crawling.rename(columns={'가게' : '지명'}, inplace=True)
```

tour_crawling의 특징, 메뉴를 word_cnt로 만들기

```python
import pandas as pd

# 메뉴와 특징 통합하여 word_cnt 생성
def combine_word_cnt(menu, features):
    try:
        # 메뉴와 특징 데이터를 문자열로 변환 후 병합
        menu_words = eval(menu) if isinstance(menu, str) else []
        feature_words = eval(features) if isinstance(features, str) else []
        # 메뉴와 특징을 통합
        combined = {word[0]: int(word[1]) for word in menu_words + feature_words if word[0]}
        return dict(sorted(combined.items(), key=lambda x: x[1], reverse=True))
    except (SyntaxError, ValueError, TypeError):
        return {}

# 메뉴와 특징을 통합하여 word_cnt 생성
tour_crawling['word_cnt'] = tour_crawling.apply(lambda x: combine_word_cnt(x['메뉴'], x['특징']), axis=1)
```
