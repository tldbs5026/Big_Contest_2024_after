변환
1. 칼럼명이 map인 경우는 int로 변환
2. object인 경우는 일단 보류
3. 칼럼명에 v(visit)인 경우도 int로 변환
4. P인 경우도 int로 변환

```python
import numpy as np
def float_to_4f(df) :
    float_cols = df.select_dtypes(include=['float64']).columns

    for col in float_cols :
        df[col] = df[col].fillna(-9)
        if col.endswith('map') :
            df[col] = df[col].astype(np.int32)
            continue
        elif col.endswith('v') :
            df[col] = df[col].astype(np.int32)
            continue
        elif col.endswith('P') :
            df[col] = df[col].astype(np.int32)
            continue


        df[col] = df[col].apply(lambda x : format(f"{x:.4f}"))
        df[col] = df[col].astype(np.float32)
    return df

convert_df = float_to_4f(data)
convert_df
```
